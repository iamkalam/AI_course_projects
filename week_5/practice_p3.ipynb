P3 The Password Strength Analyzer
FAST IT department needs a password checker. Write a function check_password(password)
that analyzes a password and returns a strength score (0-5) based on these rules: +1 for length
>= 8, +1 for containing uppercase, +1 for containing lowercase, +1 for containing a digit, +1 for
containing a special character (!@#$%^&*). Also return a list of missing criteria. Map scores to
labels: 0-1 = "Weak (your password is basically 'password123')", 2-3 = "Medium (getting
there...)", 4-5 = "Strong (even hackers would cry)". Test with at least 5 different passwords.

```python


def password_checker(password):
    score =0
    
    # Check length
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
        return False
    elif len(password) >= 8:
        score += 1
    
    # Check for uppercase letter
    if not any(char.isupper() for char in password):
        print("Password must contain at least one uppercase letter.")
        return False
    else:
        score += 1
    
    # Check for lowercase letter
    if not any(char.islower() for char in password):
        print("Password must contain at least one lowercase letter.")
        return False
    else:
        score += 1
    
    # Check for digit
    if not any(char.isdigit() for char in password):
        print("Password must contain at least one digit.")
        return False
    else:
        score += 1
    
    # Check for special character
    special_characters = "!@#$%^&*()-_=+[]{}|;:'\",.<>?/"
    if not any(char in special_characters for char in password):
        print("Password must contain at least one special character.")
        return False
    else:
        score += 1
    
    
    if score < 4:
        print("Password is weak.")
        print("your Score is", score, "out of 5.")
        return False
    elif score == 4:
        print("Password is moderate.")
        print("Your Score is 4 out of 5.")
        return True
    elif score == 5:
        print("Password is strong!")
        print("Your Score is 5 out of 5.")
        return True



print("enter a password to check its strength:")
user_password = input() 
password_checker(user_password)
    
```
