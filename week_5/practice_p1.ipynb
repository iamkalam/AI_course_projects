P1 The Chai Stall Calculator
Your university's famous chai wala (Peshawar's finest!) wants to digitize his business. He sells
chai for Rs. 50, paratha for Rs. 40, and anda (egg) for Rs. 30. Write a program that takes an
order as a dictionary (e.g., {"chai": 3, "paratha": 2, "anda": 1}) and prints a formatted receipt
showing each item, its quantity, subtotal, and the grand total. If the total exceeds Rs. 500, apply
a 10% "best customer" discount. Add a tip suggestion of 5%, 10%, and 15%.
# Expected output for order = {"chai": 3, "paratha": 2, "anda": 1}:
# ========== CHAI WALA RECEIPT ==========
# Chai x3 Rs. 150
# Paratha x2 Rs. 80
# Anda x1 Rs. 30
# ------------------------------------
# Subtotal: Rs. 260
# Discount (10%): Rs. 0
# Grand Total: Rs. 260
# ------------------------------------
# Tip Suggestions:
# 5% -> Rs. 13.00
# 10% -> Rs. 26.00
# 15% -> Rs. 39.00
# ======================================

```python
def print_receipt(order):
    # Prices
    prices = {
        "chai": 50,
        "paratha": 40,
        "anda": 30
    }
    
    # Calculate subtotals for each item
    items_output = []
    subtotal = 0
    
    for item, quantity in order.items():
        if item in prices:
            item_total = prices[item] * quantity
            subtotal += item_total
            items_output.append(f"{item.capitalize()} x{quantity} Rs. {item_total}")
    
    # Calculate discount
    discount = 0
    if subtotal > 500:
        discount = subtotal * 0.10
    
    # Calculate grand total
    grand_total = subtotal - discount
    
    # Calculate tip suggestions
    tip_5 = grand_total * 0.05
    tip_10 = grand_total * 0.10
    tip_15 = grand_total * 0.15
    
    # Print receipt
    print("========== CHAI WALA RECEIPT ==========")
    for item_line in items_output:
        print(item_line)
    print("------------------------------------")
    print(f"Subtotal: Rs. {subtotal}")
    print(f"Discount (10%): Rs. {discount}")
    print(f"Grand Total: Rs. {grand_total}")
    print("------------------------------------")
    print("Tip Suggestions:")
    print(f"5% -> Rs. {tip_5:.2f}")
    print(f"10% -> Rs. {tip_10:.2f}")
    print(f"15% -> Rs. {tip_15:.2f}")
    print("======================================")

order = {"chai": 3, "paratha": 2, "anda": 1}
print_receipt(order)
print("\n" + "="*40 + "\n")

order2 = {"chai": 5, "paratha": 5, "anda": 5}
print_receipt(order2)

```
