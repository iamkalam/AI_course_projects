P2 CGPA Analyzer: Are You Dean's List Material?
Write a program that stores 5 students' data as a list of dictionaries. Each student has a name,
roll number, and a list of 5 course GPAs. Calculate each student's CGPA (average of course
GPAs). Then: (a) Print all students sorted by CGPA (highest first), (b) Identify who made the
Dean's List (CGPA >= 3.5), (c) Find the student with the highest and lowest CGPA, (d)
Calculate the class average CGPA.

```python
# Sample data structure
students = [
    {"name": "Ali Khan", "roll": "21P-0001",
    "gpas": [3.7, 3.5, 3.8, 3.6, 3.9]},
    {"name": "Sara Ahmed", "roll": "21P-0002",
    "gpas": [3.9, 4.0, 3.7, 3.8, 3.9]},
    {"name":"Abdul kalam" , "roll" :"23P-0622",
     "gpas":[3.8, 3.9, 4.0, 3.7, 3.6]},
    
    {"name":"wajahat taj" , "roll" :"23P-3023",
     "gpas":[3.5, 4.0, 3.1, 3.7, 3.8]},
    
    {"name":"muneeb hassan" , "roll" :"22P-9199",
     "gpas":[3.2, 3.7, 3.0, 3.5, 4.0]}
]

```

```python
# Calculate CGPA for each student
for student in students:
    student["gpa"] = sum(student["gpas"]) / len(student["gpas"])
    student["cgpa"]=student["gpa"]
    
#display
print("All students with CGPA ")
print ("-" * 40)

for student in students:
    print(f"Name:{student['name']:20} | Roll: {student['roll']} | CGPA: {student['cgpa']:.2f}")
```

```python
# (a) Sort students by CGPA (highest first)
sorted_students = sorted(students, key=lambda x: x["cgpa"], reverse=True)

print("\n(a) Students Sorted by CGPA (Highest First):")
print("-" * 60)
for i, student in enumerate(sorted_students, 1):
    print(f"{i}. {student['name']:20} | CGPA: {student['cgpa']:.2f}")
```

```python
# (b) Identify Dean's List (CGPA >= 3.5)
deans_list = [student for student in students if student["cgpa"] >= 3.5]

print("\n(b) Dean's List (CGPA >= 3.5):")
print("-" * 60)
if deans_list:
    for student in deans_list:
        print(f"✓ {student['name']:20} | Roll: {student['roll']:12} | CGPA: {student['cgpa']:.2f}")
else:
    print("No students made the Dean's List")
```

```python
# (c) Find highest and lowest CGPA
highest_student = max(students, key=lambda x: x["cgpa"])
lowest_student = min(students, key=lambda x: x["cgpa"])

print("\n(c) Highest and Lowest CGPA:")
print("-" * 60)
print(f"Highest: {highest_student['name']:20} | CGPA: {highest_student['cgpa']:.2f}")
print(f"Lowest:  {lowest_student['name']:20} | CGPA: {lowest_student['cgpa']:.2f}")
```

```python
# (d) Calculate class average CGPA
class_avg_cgpa = sum(student["cgpa"] for student in students) / len(students)

print("\n(d) Class Average CGPA:")
print("-" * 60)
print(f"Class Average CGPA: {class_avg_cgpa:.2f}")
```
