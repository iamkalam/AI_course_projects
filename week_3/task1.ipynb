Task 1  PEAS Description 

scenrio : Scenario: Smart Traffic Light System


PEAS 


1. Performance Measure (What makes the agent “good”?)
    .Minimizes traffic congestion
    .Reduces total travel time for people 
    .Improves road safety
    .giving priority to emergency vechicls

2. Environment (Where does it operate?)
   .Road and vechicle density 
   .people crossing roads 
   .road constructions 
   .weather condiditons 
   .traffic laws and regulations 

3. Actuators (How does the agent output decisions?)
   .traffic lights (Red , Yellow , Green )
   .Pedestrian crossing signals
   .Digital traffic signboards showing instructions or warnings

4. Sensors (What inputs does it receive?)
   .Road cameras for monitoring vehicle queues and movement
   .Inductive loop sensors embedded in roads to detect vehicles
   .Infrared or motion sensors for pedestrian detection
   .GPS and communication data from emergency vehicles
   .Weather sensors
